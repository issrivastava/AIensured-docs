```python
from typing import List, Optional

class TreeNode:
    def __init__(self, title: str, node_type: str, content: str = ""):
        self.title = title            # e.g., "Chapter 1 : introduction" or "DOCUMENT"
        self.node_type = node_type    # "root", "section", or "leaf"
        self.content = content        # Only populated for leaf nodes (or summaries)
        self.children: List['TreeNode'] = []
        self.parent: Optional['TreeNode'] = None

    def add_child(self, child_node: 'TreeNode'):
        child_node.parent = self
        self.children.append(child_node)

    def display(self, level=0):
        """Helper to print the tree visually in the console."""
        indent = "  " * level
        content_preview = f" -> '{self.content[:30]}...'" if self.content else ""
        print(f"{indent}[{self.node_type.upper()}] {self.title}{content_preview}")
        for child in self.children:
            child.display(level + 1)
```

```python
import re

def build_document_tree(raw_text: str) -> TreeNode:
    root = TreeNode(title="DOCUMENT", node_type="root")

    # Split text by lines and clean them up
    lines = [line.strip() for line in raw_text.strip().split('\n') if line.strip()]

    current_chapter_node = None

    for line in lines:
        # Regex to detect "Chapter X : ..."
        if re.match(r'^chapter\s+\d+\s*:', line, re.IGNORECASE):
            # Create a new internal node for the chapter
            current_chapter_node = TreeNode(title=line, node_type="section")
            root.add_child(current_chapter_node)
        else:
            # It's content. Append it to the current chapter if one exists
            if current_chapter_node:
                leaf_node = TreeNode(title=f"Content for {current_chapter_node.title}", node_type="leaf", content=line)
                current_chapter_node.add_child(leaf_node)
            else:
                # Fallback if text appears before any chapter heading
                leaf_node = TreeNode(title="Preamble", node_type="leaf", content=line)
                root.add_child(leaf_node)

    return root

# --- Testing the implementation ---
sample_input = """
chapter 1 : introduction
this is the introduction part
chapter 2 : literature review
this is the literature review part
"""

tree = build_document_tree(sample_input)
tree.display()
```
